```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



df_2005 = pd.read_csv('./data/2005/merged_movies_data_2005.csv')
df_2006 = pd.read_csv('./data/2006/merged_movies_data_2006.csv')
df_2007 = pd.read_csv('./data/2007/merged_movies_data_2007.csv')
df_2008 = pd.read_csv('./data/2008/merged_movies_data_2008.csv')
df_2009 = pd.read_csv('./data/2009/merged_movies_data_2009.csv')
df_2010 = pd.read_csv('./data/2010/merged_movies_data_2010.csv')
df_2011 = pd.read_csv('./data/2011/merged_movies_data_2011.csv')
df_2012 = pd.read_csv('./data/2012/merged_movies_data_2012.csv')
df_2013 = pd.read_csv('./data/2013/merged_movies_data_2013.csv')
df_2014 = pd.read_csv('./data/2014/merged_movies_data_2014.csv')
df_2015 = pd.read_csv('./data/2015/merged_movies_data_2015.csv')
df_2016 = pd.read_csv('./data/2016/merged_movies_data_2016.csv')
df_2017 = pd.read_csv('./data/2017/merged_movies_data_2017.csv')
df_2018 = pd.read_csv('./data/2018/merged_movies_data_2018.csv')
df_2019 = pd.read_csv('./data/2019/merged_movies_data_2019.csv')
df_2020 = pd.read_csv('./data/2020/merged_movies_data_2020.csv')
df_2021 = pd.read_csv('./data/2021/merged_movies_data_2021.csv')
df_2022 = pd.read_csv('./data/2022/merged_movies_data_2022.csv')
df_2023 = pd.read_csv('./data/2023/merged_movies_data_2023.csv')
df_2024 = pd.read_csv('./data/2024/merged_movies_data_2024.csv')

df = pd.concat([df_2005, df_2006,df_2007,df_2008,df_2009,df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021,df_2022,df_2023,df_2024])

df
```

```python
df.info()
```

```python
# MPA null 값 unrated 로 바꾸기 

df['MPA'] = df['MPA'].fillna('unrated')

# Rating 열에서 null 값 평균으로 대체

rating_mean = df['Rating'].mean()
df['Rating'] = df['Rating'].fillna(rating_mean)

# votes null 값 평균으로 대체 
def convert_votes(vote_str):
    if isinstance(vote_str, str):
        if 'M' in vote_str:
            return float(vote_str.replace('M', '')) * 1000000
        elif 'K' in vote_str:
            return float(vote_str.replace('K', '')) * 1000
    return float(vote_str)
df['Votes'] = df['Votes'].apply(convert_votes)

df['Votes'] = df['Votes'].fillna(df['Votes'].mean())


```

```python
# 문자열 전처리

# 대괄호와 따옴표만 제거하고 쉼표는 그대로 두는 함수
def clean_column_values(column):
    return column.apply(lambda x: ', '.join([item.replace("'", "").replace('"', '').replace('[', '').replace(']', '') for item in str(x).split(',')]))

# 각 컬럼에 대해 전처리 수행
df_cleaned = df.apply(clean_column_values)

# 전처리된 데이터프레임 출력
print(df_cleaned)
```
