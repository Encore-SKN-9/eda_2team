```python
# MPA null 값 unrated 로 바꾸기 

df['MPA'] = df['MPA'].fillna('unrated')

# Rating 열에서 null 값 평균으로 대체

rating_mean = df['Rating'].mean()
df['Rating'] = df['Rating'].fillna(rating_mean)

# votes null 값 평균으로 대체 
def convert_votes(vote_str):
    if isinstance(vote_str, str):
        if 'M' in vote_str:
            return float(vote_str.replace('M', '')) * 1000000
        elif 'K' in vote_str:
            return float(vote_str.replace('K', '')) * 1000
    return float(vote_str)
df['Votes'] = df['Votes'].apply(convert_votes)

df['Votes'] = df['Votes'].fillna(df['Votes'].mean())


```

```python
import pandas as pd


# Duration 컬럼을 분(minute) 단위로 변환하는 함수
def str_to_minutes(duration):
    if pd.isnull(duration):
        return None  # None 값은 그대로 반환

    try:
        # 'h'가 포함된 경우, 시간과 분을 분으로 변환
        if 'h' in duration:
            duration = duration.split('h')
            hours = int(duration[0])  # 시간 값
            minutes = 0
            if 'm' in duration[1]:
                minutes = int(duration[1].split('m')[0])  # 분 값
            return hours * 60 + minutes  # 시간을 분으로 변환

        # 'm'만 있는 경우, 분 단위로 변환
        elif 'm' in duration:
            return int(duration.split('m')[0])  # 분 값만 추출

        # 숫자만 있는 경우 그대로 반환
        else:
            return int(duration)

    except Exception as e:
        return None  # 오류 발생 시 None 반환

# 'Duration' 컬럼에 str_to_minutes 함수 적용하여 분 단위로 변환
df['Duration'] = df['Duration'].apply(str_to_minutes)

# 'Duration' 컬럼을 int로 변환 (None 값은 NaN으로 대체)
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')

# 'Duration'의 평균값을 구하고, null 값을 평균값으로 채우기
mean_duration = df['Duration'].mean()

# null 값에 평균값 채우기
df['Duration'] = df['Duration'].fillna(mean_duration)

# 결과 확인
print(df[['Duration']].head())
```

```python
# 장르별 예산 결측치 처리
import pandas as pd

# 1단계: genre_budget.csv 파일을 불러옵니다.
genre_df = pd.read_csv(r'C:\qwer\solo\movie_data\data\genre_budget.csv')

# 2단계: 장르 그룹화 (비슷한 장르끼리 묶기)
genre_groups = {
    "Epic": ["Action Epic", "Fantasy Epic", "Adventure Epic", "War Epic", "Historical Epic", "Western Epic", "Romantic Epic"],
    "Fantasy": ["Fantasy", "Supernatural Fantasy", "Dark Fantasy", "Sword & Sorcery"],
    "Action": ["Action", "Car Action", "One-Person Army Action", "Martial Arts", "Gun Fu"],
    "Comedy": ["Comedy", "High-Concept Comedy", "Dark Comedy", "Slapstick", "Quirky Comedy", "Teen Comedy", "Raunchy Comedy", "Parody", "Screwball Comedy", "Stoner Comedy", "Body Swap Comedy"],
    "Horror": ["Horror", "Splatter Horror", "Slasher Horror", "Zombie Horror", "Monster Horror", "Vampire Horror", "Werewolf Horror", "Witch Horror", "Psychological Horror", "Body Horror", "Found Footage Horror", "Folk Horror", "B-Horror", "Kaiju"],
    "Romance": ["Romance", "Feel-Good Romance", "Holiday Romance", "Steamy Romance", "Tragic Romance", "Teen Romance", "Holiday Family", "Romantic Comedy"],
    "Adventure": ["Adventure", "Jungle Adventure", "Mountain Adventure", "Sea Adventure", "Desert Adventure", "Urban Adventure", "Globetrotting Adventure", "Teen Adventure"],
    "Drama": ["Drama", "Period Drama", "Legal Drama", "Crime Drama", "Medical Drama", "Political Drama", "Workplace Drama", "Teen Drama", "Psychological Drama", "Historical Drama", "Showbiz Drama", "Family Drama"]
}

# 3단계: 장르를 카테고리로 묶는 함수 정의
def categorize_genre(genre):
    for category, genres in genre_groups.items():
        if genre in genres:
            return category
    return "Other"

# 4단계: 'genres' 열을 카테고리화하여 새로운 열 'genre_category'를 추가
genre_df['genre_category'] = genre_df['genres'].apply(categorize_genre)

# 5단계: 카테고리별 예산 합계와 평균 계산
category_budget = genre_df.groupby('genre_category')['budget'].agg(['sum', 'mean']).reset_index()

# 결과를 확인
print(category_budget)

# csv 파일로 만들기 
# category_budget.to_csv('data/category_budget.csv', index=False)

# null 값에 대입하기 





```
